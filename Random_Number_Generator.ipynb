```python
import time
import random

def menu():
  # ----------------------------------------------------------------------------------------------------------------- #
  print(" ----------------------------\n Welcome to Number Guessing Game!\n ----------------------------")
  print(" Instructions:\n Enter your name then wait for a number to be generated!\n All you have to do is guess a number between 1-100!\n Have fun!\n----------------------------")
  # ----------------------------------------------------------------------------------------------------------------- #
  print("\n> Hello user, what is your name? ")
  for i in range(1,6): 
    name = input("")
    time.sleep(0.5)
    if name == " " or name == "":
      print("> Please re-enter your name. Your name cannot be blank.")
    elif name.isdigit():
      print("> Please re-enter your name. Your name cannot have numbers.")
    elif name.isalpha() == True: # want to create a program to confirm if name is correct
      print(f"\n> Welcome {name}!")
      break
    chances = 5-i
    print(f"You have {chances} attempts left  \n")
    if chances == 0:
      replay()
      
def number_generator():
      time.sleep(1)
      print("\n> Please wait for your number to generate...")
      time.sleep(4)
      print("\n> Number generated!")
      time.sleep(1.5)
      min=1
      max=101
      print("\n>> Try to guess the number! <<")
      time.sleep(1)
      print("Guess a number from 1-100!")
      gen = random.randrange(min, max) # make this a hint feature
      if gen % 2 == 0:
        even = gen
        print("(Hint: The number is even!)")
        return even
      else:
        odd = gen
        print("(Hint: The number is odd!)")
        return odd

def guessing_number():
  gen = number_generator() 
  for i in range(1,6):
    guess = input("")
    if guess == " " or guess == "":
      print("> Please enter a valid number. Your guess cannot be blank.")
    elif int(guess) < 1 or int(guess) > 100:
      print("> Please enter a number within the given range. ")
    elif int(guess) < gen:
      print("> Your number is too low!")
    elif int(guess) > gen:
      print("> Your number is too high!")
    elif int(guess) == gen:
      print("> Congratulations! You guessed the number!")
      break
    chances = 5-i
    print(f"You have {chances} attempts left  \n")
    if chances == 0:
      replay()

def replay():
  print("\n> Would you like to play again? [Yes/No] ")
  replay = input("")
  if replay == "No" or replay == "no":
      print("> Thank you for playing!")
  elif replay == "Yes" or replay == "yes":
      time.sleep(1.5)
      print("\n> Initializing game... Please wait.")
      menu()
      

menu()
guessing_number()
```
